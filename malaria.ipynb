## 2. Data Processing
- Remove columns with more than 50% missing values

```python
col_grt_50_missing = malaria.columns[((malaria.isnull().mean() * 100) < 50)]
malaria_final = malaria[col_grt_50_missing]

print(malaria_final.shape)
malaria_final.head()
```
